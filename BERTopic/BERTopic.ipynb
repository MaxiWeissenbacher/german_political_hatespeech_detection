```python
train = pd.read_csv("all_data_combined.csv")
# zum Testen nur ein paar Tweets auswählen
#train = train.head(10000)
train.head()
```

```python
train = train[["text", "created_at","Partei"]]
```

```python
# separate years
einund20 = train[train["created_at"] <= "2021-12-31 00:00:00+00:00"]
zweiund20 = train[train["created_at"] >= "2021-12-31 00:00:00+00:00"]
```

```python
train["Partei"].value_counts()
```

```python
einund20["Partei"].value_counts()
```

```python
tweets = einund20.text.tolist()
timestamps = einund20.created_at.to_list()
```

```python
from bertopic import BERTopic
```

```python
# trainieren dauert ca. eine Stunde
# Load Trained Model
topic_model = BERTopic.load("bertopic_2021")
# Train model
#topics, probs = topic_model.fit_transform(tweets)
```

```python
# include timestamps to see different topics over the whole year
topics_over_time = topic_model.topics_over_time(tweets, timestamps, nr_bins=20)
```

```python
topic_model.get_topic_info()
```

```python
%matplotlib inline

# Visualize the Plot
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=20)
```
