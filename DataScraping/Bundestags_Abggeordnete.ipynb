# Notebook to Scrape the Twitter Usernames

Scraping the Twitter Usernames with Beautifulsoup from "Bundestag.de"

```python
import requests
from bs4 import BeautifulSoup
import json
import csv
import pandas as pd
```

```python
def get_links():
    person_url_list = []
    for i in range(0,740,20):
        url = f'https://www.bundestag.de/ajax/filterlist/de/abgeordnete/862712-862712?limit=20&noFilterSet=true&offset={i}'
        q = requests.get(url)
        result = q.content
        soup = BeautifulSoup(result,"lxml")
        persons = soup.find_all(attrs={'class': 'bt-slide-content'})
        for person in persons:
            person_page_url = person.find('a')['href']
            person_url_list.append(person_page_url)
            
    with open("person_url_list.txt", "a") as file:
            for line in person_url_list:
                file.write(f"https://www.bundestag.de{line}\n")
```

```python
get_links()
```

```python
# Scrape all the needed information and save it in a dataframe
def get_infos():
    with open("person_url_list.txt") as file:
        lines = [line.strip() for line in file.readlines()]
    #with open("person_url_list.txt") as file:
     #   lines = [line.strip() for line in file.readlines()]
        overall_data = []
        counter = 0
        twitter = "HA"
        for line in lines:
            counter = counter + 1
            print(f"Line Nummer {counter}")
            q = requests.get(line)
            results = q.content
            soup = BeautifulSoup(results,"lxml")
            person = soup.find(class_= "bt-biografie-name").find("h3").text
            person_name_party = person.strip().split(",")
            person_name = person_name_party[0]
            person_party = person_name_party[1]
            print(person_name)
            social_networks = soup.find_all(class_ = "bt-link-extern")
            social_network_urls = []
            
            for social_network in social_networks:
                twitter_available = social_network.get("title") == "Twitter"
                if twitter_available==True:
                    twitter = social_network.get("href")
                    social_network_urls.append(twitter)
                    print(social_network_urls)

                
            data = {
                "full_name": person_name,
                "party": person_party,
                #"social_networks": social_network_urls,
                "twitter": social_network_urls,
            }
            

            overall_data.append(data)
            df = pd.DataFrame(data=overall_data)
        df.to_csv("scraped_twitter_names.csv")
        print("Dataframe saved on local machine")

       # with open("data.json", "w", encoding="utf-8") as json_file:
        #    json.dump(overall_data, json_file, indent=4, ensure_ascii=False)
```

```python
# It can be seen, that not every person has an Twitter Link
get_infos()
```

```python
# Extract just the Twitter Usernames from the Link in column "twitter"

username_list = []

for index, row in df.iterrows():
    if row["twitter"] != "[]":
        username = row['twitter'].rsplit('/', 1)[1].rsplit('?',1)[0].rsplit("']",1)[0]
        #print(username)
    #row["username"] = username
    else:
        username = "NA"
    username_list.append(username)
df["username"] = username_list
```
